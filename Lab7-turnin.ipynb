```python
# Exercise 13
birthrate = srange(0.1,1.5,0.1)
var("N, P")
amplitudes = []
for b in birthrate :
    t = srange(0,100,0.1)
    sol=desolve_odeint([b*N - 0.01*N*P, 0.5*0.01*N*P - 0.2*P], ics=[100,200], dvars=[N,P], times=t)
    amplitudes.append((max(sol[:,1])-min(sol[:,1]))/2.0)
list_plot(zip(birthrate,amplitudes), plotjoined =true)
```
